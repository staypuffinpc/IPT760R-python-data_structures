```python
#first, let's connect to our data
import pymongo

connection = pymongo.MongoClient("ds031925.mlab.com",31925)
db = connection["f16_class_demo"]
user = "760R_student"
pwd = "justforfun"
db.authenticate(user, pwd) # if this works, you should get a True result, False otherwise
```

```python
#let's start by dumping all the data into a single list
posts = []
for x in db.log.find():
    posts.append(x)
```

```python
#create a "data frame" for viewing and messing around with the data
import pandas as pd
df = pd.DataFrame(posts)

#just expressing the dataframe will print it out in a spreadsheet-like format
df
```

```python
#see what your fields are
df.columns
```

```python
#describe the quantitative data
df.describe()
```

## Filter and Sort data

```python
#find a single record
# df.head(5)
df.tail(3)
# df.loc[100]
```

```python
#find records with an age between 20 and 30 (including 20 and 30)
df[(df['age']<=30) & (df['age']>= 20)]
```

```python
#find all posts where the author's name is NOT Peter 
df[df.author == "Peter"]
```

```python
#find all posts by Greg OR by Marcia.  Notice the use of the pipe "|" symbol for the "or" statement
df[(df.author == 'Greg')|(df.author == 'Marcia')]
```

```python
#find all posts that have a "second_author" field
df[df.second_author == df.second_author]
```

```python
#sorting
df.sort_values('age',ascending=False)
```

```python
df[df.author == 'Peter'].sort_values('age',ascending=False)
#find all posts by author: Peter, then sort those based on the 'age' field
```

```python
#do a word count on titles to see frequency
words = []
for x in df.title:
    if x != x:
        continue
    for y in x.split(' '):
        words.append(y)

pd.Series(words).value_counts()
```

```python
#find posts with only 2 tags
df[df.tags.map(lambda x: len(x) if x == x else 1) == 2]
```

## Messing with Numbers

```python
#first, get the needed libraries for dealing with numbers
import numpy as np
import math
```

```python
#get the average age (mean, median, then mode)
df['age'].mean()
```

```python
df['age'].median()
```

```python
df['age'].mode()
```

```python
#find the mean age of posts by author
df.groupby('author').mean()
```

```python
#count the number of records for each author
counts = df.groupby('author').count()
counts.dropna()
```

```python
#iterate over each item in a series and subtract 2 (lambda creates an anonymous function in python)
df['age'].map(lambda a:a-2)
```
