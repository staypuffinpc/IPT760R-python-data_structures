```python
# Demonstration of how to read and write to XML files using Python

#if not familiar with XML, please complete the tutorial at http://www.w3schools.com/xml/

import xml.etree.ElementTree as ET # tree gets the entire file
tree = ET.parse('states.xml')
root = tree.getroot()

print (root.tag) # to find out the name of the root
print (root.attrib) #for root attributes as a dict
```

```python
#iterate over the children of root
for child in root:
    print (child.tag, child.attrib)
```

```python
# access a specific node within root
print (root[5][0].text, root[5][1].text)
```

```python
#to iterate over the entire tree recursively (i.e., dig down in all its levels), use .iter() method
for l in root.iter('Location'):
    print (l[0].text, l[0].attrib, l[1].text, l[2].attrib)
```

```python
for state in root.findall('Location'):
    abbr = state.find('abbreviation').text
    name = state.find('state').text
    print (name, "("+abbr+")")
```

```python
#to add (append) a new field to a node
for loc in root.findall('Location'):
    print (list(loc)) #to verify, show all this Element's children
```

```python
#create a number and assign it as a value to the Population tag
import random
random.seed()
root = tree.getroot()
for loc in root.findall('Location'):
    num = random.randint(10000,1000000)
    loc.find('population').text = str(num)
    print (loc.find('state').text, ", Population: ",loc.find('population').text)
```

```python
#use XPATH notation to find nodes
for loc in root.findall('*/'):
    print (loc.tag, loc.attrib,"\n")
```

```python
#make sure you make the changes to the file permament
tree.write('output.xml')
```
