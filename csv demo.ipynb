```python
import csv

#we'll open a file with some data from the last U.S. Census and leave it open for the majority of these examples
data = "NST-EST2014-01.csv"
f = open(data,"rU")  #adding the U to the read type allows it to be open for universal reading.  This gets rid of the "newline" character problem

#create a list to which you'll save this info
def readCSV():
    try:
        reader = csv.reader(f)
        new_data = []
        for row in reader:
            new_data.append(row)
    except csv.Error, e:
        print('ERROR: file %s, line %d: %s' % (data, reader.line_num, e))
    finally:
        f.close()
        return new_data
thisData = readCSV()
print thisData
```
